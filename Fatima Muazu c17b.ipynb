```python
#dictionary for products inventory
beverageDict={'chocolate':7,'drinks':50,'coffee':25,'tea':18,'soyDrinks':35,'pop':20,'soda':15}
phoneAccDict={'c_case':12,'earpieces':15,'s_guards':10}
toiletDict={'t_taper':25,'soap':8,'scrubs':5,'cream':20,'shampoo':15}
pastryDict={'pizza':20,'burgers':15,'donuts':25,'muffins':18,'cheesecakes':12}
cosmeticDict={'perfumes':23,'vanishes':8,'nail_polish':7,'deodorants':25,'face_scrub':32}
```

```python
#checking inventory for availability
def checkInventory(product, quantity):
    if product in beverageDict:
        print('product entered is a beverage')
    else:
        print('product is not in the beverage inventory')
    for k,v in beverageDict.items():
        if v<5:
             print('our stock is less than 5')
        else:
             v-=quantity
        print(f'the number of {k} left is {v}')
        
    if product in phoneAccDict:
        print('product entered is a phone accessory')
    else:
        print('product is not in the phone accessory inventory')
    for k,v in phoneAccDict.items():
        if v<5:
             print('our stock is less than 5')
        else:
             v-=quantity
        print(f'the number of {k} left is {v}')  
    
    if product in toiletDict:
        print('product entered is a toiletry')
    else:
        print('product is not in the toiletry inventory')
    for k,v in toiletDict.items():
        if v<5:
             print('our stock is less than 5')
        else:
             v-=quantity
        print(f'the number of {k} left is {v}')
        
    if product in pastryDict:
        print('product entered is a pastry')
    else:
        print('product is not in the pastry inventory')
    for k,v in pastryDict.items():
        if v<5:
             print('our stock is less than 5')
        else:
             v-=quantity
        print(f'the number of {k} left is {v}')
    
    if product in cosmeticDict:
        print('product entered is a cosmetic')
    else:
        print('product is not in the cosmetics inventory')
    for k,v in cosmeticDict.items():
        if v<5:
             print('our stock is less than 5')
        else:
             v-=quantity
        print(f'the number of {k} left is {v}')
        
```

```python
product=input('enter a product here ')
quantity=int(input('enter quantity for purchase'))
result=checkInventory(product, quantity)
```
