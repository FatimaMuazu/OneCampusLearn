```python
#in data modelling, the input features are used to predict the
#outcome feature. it is always possible that some of the input features
#may not be contirbuting significantly to predicting the outcome fetaure
#there are also scenarios where some features may not be contributing at all.

#features that only contribute little or nothing to the prediction can 
#constitute noise and make our models perform poorly.
#it is therefore important to conduct feature selection to identify 
#those features that do not contribute and eliminate them 
#
```

```python
#UNIVARIATE focuses on one feature at a time through inpecting the
#strength of one feature at a time
#it checks the strenght of each feature in directly influencing
#the outcome variable using CHI-SQUARE test.

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#convert dataframe to array
dfArr = df.values
X = dfArr[:,:-1]
Y = dfArr[:,-1]

#Rule: whenever you are trying to select features in your data,
#use half of your feature therefore k =4 to get the top 4 features
#with the highest strengths
test = SelectKBest(score_func=chi2, k=5)
#fitting is a process of finding a line of best fit
#gradient of the line represent coefficeint factor
fit = test.fit(X,Y)
features = fit.transform(X)
```

```python
fit.scores_
```

```python
#we can build a model 
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
res = model.fit(X,Y)
score = res.score(X,Y)

```

```python
#RECURSIVE FEATURE ELIMINATION
#it checks the effect between a feature and the outcome by removing one feature, train
#the model then put that feature back, then remove the next one, train the model,
#and put it back.
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
dfArr = df.values
X = dfArr[:,:-1]
Y = dfArr[:,-1]

rfe = RFE(model, n_features_to_select= 5)
fit = rfe.fit(X,Y)
```
